Terdapat beberapa perbedaan karakteristik pada dua data diatas yaitu:
1. Jumlah Data yang Valid (count)
Pada data train features, kolom seperti tahun_kelahiran memiliki 3817 entri valid, sementara pada pada data test features, jumlah entri valid untuk kolom yang sama adalah 3818.
Perbedaan ini menunjukkan bahwa dataset di kedua data memiliki satu baris tambahan yang valid untuk beberapa kolom.

2. Nilai mean (Rata-rata)
Ada perbedaan dalam nilai rata-rata untuk beberapa kolom. Pendapatan pada data train features memiliki rata-rata 1.144832e+08, sedangkan pada data test features rata-rata adalah 1.163249e+08. Hal ini menunjukkan adanya perbedaan dalam nilai data antara kedua dataset tersebut, mungkin karena penambahan atau penghapusan entri tertentu.

3. Nilai std (Standar Deviasi)
Standar deviasi di beberapa kolom juga berbeda. Misalnya, pada kolom belanja_daging, standar deviasi pada dataset train features adalah 5.120427e+05, sedangkan pada dataset test features adalah 5.123206e+05.
Ini menunjukkan bahwa penyebaran data di kedua dataset sedikit berbeda.

4. Nilai Minimum (min) dan Maksimum (max)
Beberapa nilai minimum dan maksimum antara kedua datasdet tidak sama. Misalnya:
Pada kolom belanja_daging, nilai maksimum pada dataset train features  adalah 3.489675e+06, sedangkan pada dataset test features adalah 3.271191e+06. Ini mengindikasikan bahwa terdapat perbedaan nilai ekstrim di antara dataset yang dibandingkan.

5. Percentile (25%, 50%, 75%)
Perbedaan kecil juga terjadi pada nilai kuartil pertama (25%), median (50%), dan kuartil ketiga (75%). Pada
Kolom belanja_buah memiliki nilai median 26.456 pada dataset train features, sementara pada dataset test features nilai median adalah 29.525.
Ini menunjukkan perbedaan distribusi data di kedua dataset.

```python
num_cols = [
    'tahun_kelahiran',
    'pendapatan',
    'belanja_buah',
    'belanja_daging',
    'belanja_ikan',
    'belanja_kue',
    'pembelian_diskon'
]
```

```python
corr = features[num_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Pada korelasi diatas yang merupakan representasi visual dari koefisien korelasi antara berbagai variabel dalam suatu dataset. Variabel-variabel tersebut tercantum di sumbu x dan y, dan nilai-nilai dalam gambar diatas mewakili koefisien korelasi antara variabel-variabel yang bersesuaian.

Koefisien korelasi berkisar antara -1 hingga 1, di mana -1 menunjukkan korelasi negatif yang kuat, menunjukkan tidak ada korelasi, dan 1 menunjukkan korelasi positif yang kuat. pada gambar diatas menggunakan skala warna untuk mewakili nilai-nilai korelasi yang berbeda, dengan biru tua menunjukkan korelasi positif yang lebih kuat dan merah tua menunjukkan korelasi negatif yang lebih kuat.

Sebagai contoh, koefisien korelasi antara "belanja_daging" dan "belanja_daging" adalah 1,00, yang menunjukkan korelasi positif sempurna antara dua variabel ini. Di sisi lain, koefisien korelasi antara "tahun_kelahiran" dan "pendapatan" adalah -0,19, yang menunjukkan korelasi negatif yang lemah antara dua variabel ini.
